```python
from pyspark.sql import SparkSession
spark = SparkSession \
    .builder \
    .master('local[*]') \
    .config("spark.driver.memory", "11g") \
    .appName('my-cool-app') \
    .getOrCreate()

sc=spark.sparkContext
```

```python
import re
deleted_words = ['' ]
def clean_text(str1):
    for word in deleted_words:
        str1= str1[:len(str1)-1]
        str1 = str1.replace(" " , ',')

    return str1

def make_unique(str1):
    temp = ''
    temp_str = str1.split(',')
    sec_temp = []
    for word in temp_str:
        if word in sec_temp:
            pass
        else:
            sec_temp.append(word)
            temp += word + ','
    return temp[:len(temp)-1]
    
```

```python
file = sc.textFile("records.txt").map(clean_text).map(make_unique)
file.collect()
```

```python
lblitems = file.map(lambda line: line.split(','))
lblitems.collect()
```

```python
wlitems = file.flatMap(lambda line:line.split(','))
print(len(wlitems.collect()))
wlitems.collect()
```

```python
from operator import add

uniqueItems = wlitems.distinct()
supportRdd = wlitems.map(lambda item: (item , 1))
supportRdd = supportRdd.reduceByKey(add)

supportRdd.sortBy(lambda x:-x[1]).collect()

```

```python
supportRdd = supportRdd.filter(lambda item: item[1] >= 100 )
```

```python
baseRdd = supportRdd.map(lambda item: ([item[0]] , item[1])) 

supportRdd = supportRdd.map(lambda item: item[0])
supportRdd.collect()
```

```python
def removeReplica(record):

    if(isinstance(record[0], tuple)):
        x1 = record[0]
        x2 = record[1]
    else:
        x1 = [record[0]]
        x2 = record[1]

    if(any(x == x2 for x in x1) == False):
        a = list(x1)
        a.append(x2)
        a.sort()
        result = tuple(a)
        return result 
    else:
        return x1
```

```python
c = 2 # Combination length 
frequent_items = []
while(c <= 3):

    combined = supportRdd.cartesian(uniqueItems)
    combined = combined.map(lambda item: removeReplica(item))
  
    combined = combined.filter(lambda item: len(item) == c)
    combined = combined.distinct()

    
    combined_2 = combined.cartesian(lblitems)
    combined_2 = combined_2.filter(lambda item: all(x in item[1] for x in item[0]))
    
    combined_2 = combined_2.map(lambda item: item[0])
    combined_2 = combined_2.map(lambda item: (item , 1))
    combined_2 = combined_2.reduceByKey(add)
    combined_2 = combined_2.filter(lambda item: item[1] >= 100)
    frequent_items.append(combined_2.collect())
    supportRdd= combined_2.map(lambda item: item[0])
    c+=1
```

```python

from pyspark.sql import SparkSession
spark = SparkSession \
    .builder \
    .master('local[*]') \
    .config("spark.driver.memory", "11g") \
    .appName('my-cool-app') \
    .getOrCreate()

sc=spark.sparkContext
```

```python
def generate_next_c(f_k, k):
    next_c = [var1 | var2 for index, var1 in enumerate(f_k) for var2 in f_k[index + 1:] if
              list(var1)[:k - 2] == list(var2)[:k - 2]]
    return next_c


def generate_f_k(sc, c_k, shared_itemset, sup):
    def get_sup(x):
        x_sup = len([1 for t in shared_itemset.value if x.issubset(t)])
        if x_sup >= sup:
            return x, x_sup
        else:
            return ()

    f_k = sc.parallelize(c_k).map(get_sup).filter(lambda x: x != ()).collect()
    return f_k
```

```python
# read the raw data
file = sc.textFile("records.txt").map(clean_text).map(make_unique)
file.collect()# count the total number of samples
lblitems = file.map(lambda line: line.split(','))
n_samples = file.count()
# min_sup to frequency
sup = 100
# split sort

# share the whole itemset with all workers
shared_itemset = sc.broadcast(lblitems.map(lambda x: x).collect())
# store for all freq_k
frequent_itemset = []

# prepare candidate_1
k = 1
c_k = lblitems.flatMap(lambda x: set(x)).distinct().collect()
c_k = [{x} for x in c_k]
c_k
# when candidate_k is not empty
while k < 3:
        # generate freq_k
        f_k = generate_f_k(sc, c_k, shared_itemset, sup)

        frequent_itemset.append(f_k)
        # generate candidate_k+1
        k += 1
        c_k = generate_next_c([set(item) for item in map(lambda x: x[0], f_k)], k)
frequent_itemset[0]
```
